```python
%matplotlib notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = np.random.randint(0, 2, (10,10))
b = np.empty((0, 2))
print(a)


for i, n in enumerate(a):
    for j, m in enumerate(n):
        if m == 1:
            b = np.append(b, np.array([[i, j]]), axis = 0)

fig = plt.figure()
plt.gca().set_aspect('equal', adjustable="box")
plt.scatter(b[:, 0], b[:,1], color = "black", marker="d", linewidth = 10)
plt.show()
print(b)
```
